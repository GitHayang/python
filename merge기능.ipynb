```python
import pandas as pd
```

```python
masterData = pd.read_csv("../dataset/joindata1.csv")
masterData
```

```python
masterData["new_sutatin"] = masterData.suta_tin
masterData
```

```python
subData = pd.read_csv("../dataset/joindata2.csv", encoding='ms949')
subData
```

```python
key = ['mata_tin','mate_srcs_cntn','rit_strt_dt','rit_end_dt','suta_tin']
```

```python
pd.merge(left=masterData, right=subData,
         on = key,
         how='inner')
```

```python
pd.merge(left=masterData, right=subData,
         on = key,
         how='left')
```

```python
pd.merge(left=masterData, right=subData,
         on = key,
         how='right')
```

```python
pd.merge(left=masterData, right=subData,
         on = key,
         how='outer')
```

```python
pd.merge(left=masterData, right=subData,
         left_on = ['suta_tin'], right_on = ['suta_tin'],
         how='left')
```

```python
pd.merge(left=masterData, right=subData,
         left_on = ['suta_tin'], right_on = ['suta_tin'],
         how='inner')
```

```python
pd.merge(left=masterData, right=subData,
         left_on = ['suta_tin'], right_on = ['suta_tin'],
         how='right')
```
