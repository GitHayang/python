```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
featureData = pd.read_csv("../../1학기/dataset/kopo_decision_tree_all_new.csv")
featureData.head()
```

```python
groupKey = ["PRODUCTGROUP", "ITEM"]
```

```python
groupData = featureData.groupby(by=groupKey)[["YEARWEEK"]].agg("count").reset_index()
groupData = groupData.rename(columns={"YEARWEEK":"YEARWEEK_CNT"})
groupData
```

```python
mergedData = pd.merge(left = featureData, right = groupData, on = groupKey)
mergedData
```

```python
stdCnt = 50
```

```python
mlTargetData = mergedData.loc[mergedData.YEARWEEK_CNT > stdCnt].reset_index(drop = True)
mlTargetData
```

```python
ynle = LabelEncoder()
```

```python
mlTargetData["PROMOTION_YN"] = ynle.fit_transform(mlTargetData["PROMOTION"])
```

```python
yn2le = LabelEncoder()
```

```python
mlTargetData["HOLIDAY_YN"] = ynle.fit_transform(mlTargetData["HOLIDAY"])
```

```python
mlTargetData
```

```python
mlTargetData.corr()
```

```python
groupData = mlTargetData.groupby(by=groupKey)

eachGroup = groupData.get_group(list(groupData.groups)[0])
eachGroup.head()
```

```python
corrDf = eachGroup.corr()
corrDf
```

```python
stdCorr = 0.5
```

```python
label = ["QTY"]
```

```python
feature = list(corrDf.loc[(abs(corrDf["QTY"] >= stdCorr)) & (abs(corrDf["QTY"] != 1))].index)
feature
```

```python
train_feat,\
test_feat,\
train_label,\
test_label = train_test_split(eachGroup.loc[:,feature], eachGroup.loc[:,label], test_size = 0.3)
```

```python
print(train_feat.shape)
print(train_label.shape)
print(test_feat.shape)
print(test_label.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
modelDt = DecisionTreeRegressor(random_state=10)
```

```python
modelDtFit = modelDt.fit(X = train_feat, y = train_label)
```

```python
modelDtPred = modelDtFit.predict(test_feat)
modelDtPred
```

```python
testIndex = test_feat.index
testIndex
```

```python
testDataAll = eachGroup.loc[testIndex]
```

```python
testDataAll["PREDICT"] = modelDtPred
```

```python
def mlGroup(eachData):
    
    print(eachGroup.loc[:,groupKey].head(1))
    
    # groupData = mlTargetData.groupby(by=groupKey)
    # eachGroup = groupData.get_group(list(groupData.groups)[0])
    # eachGroup.head()

    corrDf = eachGroup.corr()
    corrDf
    
    startCorr = 0.5
    endCorr = 0.2

    label = ["QTY"]

    for i in range(int(startCorr*10), int(endCorr*10) - 1, -1):
        current_corr = i / 10
        feature = list(corrDf.loc[(abs(corrDf["QTY"]) >= current_corr) & (abs(corrDf["QTY"]) != 1)].index)
        print(f'Current Corr: {current_corr}, Feature List: {feature}')

    
        train_feat,\
        test_feat,\
        train_label,\
        test_label = train_test_split(eachGroup.loc[:,feature], eachGroup.loc[:,label], test_size = 0.3)

        print(train_feat.shape)
        print(train_label.shape)
        print(test_feat.shape)
        print(test_label.shape)

        from sklearn.tree import DecisionTreeRegressor

        modelDt = DecisionTreeRegressor(random_state=10)

        modelDtFit = modelDt.fit(X = train_feat, y = train_label)

        modelDtPred = modelDtFit.predict(test_feat)
        modelDtPred

        testIndex = test_feat.index
        testIndex

        testDataAll = eachGroup.loc[testIndex]

        testDataAll["PREDICT"] = modelDtPred

        return testDataAll
```
