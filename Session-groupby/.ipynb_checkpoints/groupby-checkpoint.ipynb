### 2.데이터 불러오기

### 3.머신러닝 적용성 체크

```python
### 머신러닝을 50건 이상인 아이템에 대해서만 돌리겠다.
```

```python
# pk를 정확히 확인해야한다.
groupKey = ["REGIONID", "ITEM"] 
```

```python
groupData = featureData.groupby(by = groupKey)[["YEARWEEK"]].agg("count").reset_index()
groupData.rename(columns={"YEARWEEK": "YEARWEEK_CNT"}, inplace=True)
groupData
```

```python
### 데이터 병합 (기존데이터 + 카운트 데이터)
```

```python
mergedData = pd.merge(left = featureData, right= groupData, on=groupKey)
mergedData
```

```python
mlActive = 50
```

```python
mlTargetData = mergedData.loc[mergedData.YEARWEEK_CNT > mlActive]
```

```python
mergedData.drop_duplicates(groupKey).shape
```

```python
mlTargetData.drop_duplicates(groupKey).shape
```

### 4-2.특성선정 groupby

```python
def mlGroup(eachData):
    print(eachGroup.loc[:,groupKey].head(1))
    # groupData = mlTargetData.groupby(by=groupKey)

    # eachGroup = groupData.get_group(list(groupData.groups)[2])
    eachGroup

    corrDf = eachGroup.corr()

    # for문으로 start와 end를 돌려서 feature를 끄집어내서 만든다.
    stdCorr = 0.2

    label = ["QTY"]

    features = list(corrDf.loc[(corrDf.QTY > stdCorr) & (corrDf.QTY != 1)].index)
    features

    ### 4-3.데이터 분리

    from sklearn.model_selection import train_test_split

    train_feat,\
    test_feat,\
    train_label,\
    test_label= train_test_split(eachGroup.loc[:,features],
                                 eachGroup.loc[:,label],
                                 test_size=0.2)

    print(train_feat.shape)
    print(train_label.shape)
    print(test_feat.shape)
    print(test_label.shape)

    ### 4-4.데이터 머신러닝 적용

    from sklearn.tree import DecisionTreeRegressor

    modelDt = DecisionTreeRegressor(random_state=10)

    modelDtFit = modelDt.fit(X = train_feat, y = train_label)

    ### 4-5.

    pradictValue = modelDtFit.predict(test_feat)

    testIndex = test_feat.index

    testDataAll = eachGroup.loc[testIndex]

    testDataAll["PRADICT_DT"] = pradictValue

    return testDataAll
```

```python
groupData = mlTargetData.groupby(by=groupKey).apply(mlGroup)
groupData
```
