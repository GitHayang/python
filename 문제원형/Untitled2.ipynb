```python
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import types
import pandas as pd
import os
import cv2
```

```python
df1 = pd.read_csv("./22_1.csv")
df1
```

```python
inId = "root"
inPw = "0000"
inIp = "127.0.0.1"
inPort = "3306"
inDbName = "web"
```

```python
engine = create_engine("mysql+pymysql://{}:{}@{}:{}/{}".format(inId,inPw,inIp,inPort,inDbName))

tableName = "TEST_3"
tableName = tableName.upper()
#### 문자컬럼에 대해서 varchar (100) 적용 _ 사용시 속도 SpeedUp 50배 
objectColumns = list(df1.columns[df1.dtypes == 'object'])
typeDict={}
maxLen = 100

for i in range(0, len(objectColumns)):
    # sawonData[i].str.len().max() 최대치 사용할 경우
    typeDict[ objectColumns[i] ] = types.VARCHAR(100)
#### 문자컬럼에 대해서 varchar (100) 적용 _ 사용시 속도 SpeedUp 50배 

df1.to_sql(name=tableName, if_exists="replace", con=engine,dtype=typeDict, index=False)
```
