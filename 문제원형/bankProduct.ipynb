```python
import requests
import pprint
import json
import pandas as pd
```

```python
# url 입력
url = 'https://www.fss.or.kr/openapi/api/rpGuaranteedProdList.json?key=f9f2f268874664edb46311830dd7834d&areaCode=1&sysType=1&productType=1&reportDate=2020/07'
```

```python
# url 불러오기
response = requests.get(url)
response
```

```python
#데이터 값 출력해보기
contents = response.text
contents
```

```python
# 데이터 결과값 예쁘게 출력해주는 코드
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
#문자열을 json으로 변경
json_ob = json.loads(contents)
print(json_ob)
print(type(json_ob)) #json타입 확인
```

```python
# 필요한 내용만 꺼내기
body = json_ob['list']
print(body)
```

```python
# Dataframe으로 만들기
dataframe = pd.json_normalize(body)
print(dataframe)
```

```python
df1 = pd.DataFrame(dataframe)
df1
```
