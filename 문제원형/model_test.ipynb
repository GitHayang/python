```python
import pandas as pd
import pickle
```

```python
# pickle 파일을 읽기
with open("../test/tree_model.pickle", "rb") as fr:
    tree_model = pickle.load(fr)
```

```python
### 모든 경우의 수 실행
info = '{},{},{},{},{}'
resultList = []
conditionList = []

for i in range(0, 2):
    for j in range(0, 11):
        for k in range(0, 2):
            for h in range(1, 6):
                for l in range(0, 2):
                    formatted_info = info.format(i, j, k, h, l)
                    conditionList.append(formatted_info)
                    values = [float(val) for val in formatted_info.split(',')]
                    test1 = pd.DataFrame([values])
                    predict = tree_model.predict(test1)
                    resultList.append(predict)
```

```python
resultDf1 = pd.DataFrame(resultList, columns=["PRODUCT"])
```

```python
resultDf2 = pd.DataFrame(conditionList, columns=["조건"])
```

```python
resultDf = pd.concat([resultDf2, resultDf1], axis=1)
```

```python
# row 생략 없이 출력
pd.set_option('display.max_rows', None)
resultDf.drop_duplicates().reset_index(drop=True)
```
