```python
# 라벨부분 컬럼
expandColumn = ['ind_ahor_fin_ult1', 'ind_cno_fin_ult1',\
                'ind_ctju_fin_ult1', 'ind_deco_fin_ult1', 'ind_deme_fin_ult1',\
                'ind_dela_fin_ult1', 'ind_fond_fin_ult1', 'ind_plan_fin_ult1',\
                'ind_pres_fin_ult1', 'ind_tjcr_fin_ult1', 'ind_valo_fin_ult1']
```

```python
# 컬럼
dataCol = ['sexo', 'age', 'indrel_1mes', 'tiprel_1mes', 'indresi', 'indext', 'renta', 'segmento']
```

```python
import tqdm
```

```python
### 파일 불러오기
```

```python
oneStepDf = pd.read_csv("../divide/data1.csv")
```

```python
wholeList = []
for i in tqdm.tqdm(range(len(oneStepDf))):
    for j in expandColumn:
        rowList = []
        rowList.append(oneStepDf[j][i])
        for k in dataCol:
            rowList.append(oneStepDf[k][i])
        wholeList.append(rowList)
secondStepDf = pd.DataFrame(wholeList)
```

```python
secondStepDf = secondStepDf.rename(columns={0:"product"})
secondStepDf = secondStepDf.rename(columns={1:"sexo"})
secondStepDf = secondStepDf.rename(columns={2:"age"})
secondStepDf = secondStepDf.rename(columns={3:"indrel_1mes"})
secondStepDf = secondStepDf.rename(columns={4:"tiprel_1mes"})
secondStepDf = secondStepDf.rename(columns={5:"indresi"})
secondStepDf = secondStepDf.rename(columns={6:"indext"})
secondStepDf = secondStepDf.rename(columns={7:"renta"})
secondStepDf = secondStepDf.rename(columns={8:"segmento"})

secondStepDf
```
