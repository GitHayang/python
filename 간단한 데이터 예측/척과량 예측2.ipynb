```python
#데이터 처리 라이브러리
import pandas as pd
import numpy as np

#머신러닝 라이브러리
from sklearn import tree
```

```python
basicData = pd.read_csv("../dataset/train.csv")
basicData
```

```python
totalCol = list(basicData)
targetColIndex = []
for i in range(len(totalCol)):
    targetColIndex.append(i)
    if totalCol[i] == "수관폭평균" :
        break
targetColIndex
```

```python
featuredData = basicData.iloc[:,targetColIndex]
featCol = list(featuredData.columns)
featCol
```

```python
changeCol = []
for i in range (len(featCol)):
    splitCol = featCol[i].split("(")
    changeCol.append(splitCol[0])
    featuredData.rename(columns={featCol[i]:changeCol[i]}, inplace=True)
featuredData
```

```python
featuredData.isnull().sum()
```

```python
featuredData.corr()
```

```python
featCol = list(featuredData.columns)
label = ["착과량"]
feature = featCol[2:len(featCol)]
print(label)
print(feature)
```

```python
stdRatio = 0.7
```

```python
stdIndex = int(featuredData.shape[0] * stdRatio)
stdIndex
```

```python
trainData = featuredData.loc[0:stdIndex, :]
testData = featuredData.loc[stdIndex+1:, :]
```

```python
# 70
trainDataFeatures = featuredData.loc[0:stdIndex, feature]
trainDataLabel = featuredData.loc[0:stdIndex, label]
# 30
testDataFeatures = featuredData.loc[stdIndex+1:, feature]
testDataLabel = featuredData.loc[stdIndex+1:, label]
```

```python
dtModel = tree.DecisionTreeRegressor(random_state=10)
dtModel
```

```python
trainDataFeatures
```

```python
dtModel.fit(X = trainDataFeatures,
            y = trainDataLabel)
```

```python
predictDT = dtModel.predict(X = testDataFeatures)
```

```python
testDataLabel["DT"] = predictDT
testDataLabel
```
