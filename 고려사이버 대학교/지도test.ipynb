```python
m = folium.Map(location=[loc_df.latitude[0], loc_df.longitude[0]], zoom_start=12)

# 모든 좌표에 대해 반복하여 마커를 지도에 추가합니다.
for i in range(loc_df.shape[0]):
    folium.Marker([loc_df.latitude[i], loc_df.longitude[i]], popup=(loc_df.docity[i])).add_to(m)

m
```
