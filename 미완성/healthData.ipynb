-----------혈압

```python
hyulApData = pd.read_csv("./dataset/hyulApData.csv")
hyulApData
```

```python
plt.figure(figsize=(20,5))
plt.bar(hyulApData.get_time, hyulApData.systolic, label='systolic')
plt.bar(hyulApData.get_time, hyulApData.diastolic, label='diastolic')

# 그래프 제목과 레이블 설정
plt.title('relation of systolic, diastolic')
plt.xlabel('date')
plt.ylabel('target')

# 범례 추가
plt.legend()

# x축 글자 회전
plt.xticks(rotation=90)

# 그래프 표시
plt.show()
```

```python
def date(hyulApData):
    hyulApData["date"] = 0
    for i in range(0,len(hyulApData)):
        hyulApData["date"][i] =  hyulApData["get_time"][i][:13]
```

```python
plt.figure(figsize=(20,5))
plt.bar(mergedData.date, mergedData.glucosedata)
plt.xticks(rotation=90)
plt.xlabel('date')
plt.ylabel('glucosedata')
plt.show()
```
