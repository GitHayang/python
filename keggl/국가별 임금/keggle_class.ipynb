```python
# copy()를 사용해서 복사하지 않으면 원본과 복사본 전부 값이 바뀐다.
encodeData = baseData.copy()
```

```python
# LabelEncoder()로 숫자 컬럼으로 변경
labelEn = LabelEncoder()

encodeData["work_year2"] = labelEn.fit_transform(encodeData.work_year)
encodeData.drop(columns=["work_year2"])

encodeData["salary_currency2"] = labelEn.fit_transform(encodeData.salary_currency)
encodeData.drop(columns=["salary_currency2"])

encodeData["experience_level2"] = labelEn.fit_transform(encodeData.experience_level)
encodeData.drop(columns=["experience_level2"])

encodeData["employment_type2"] = labelEn.fit_transform(encodeData.employment_type)
encodeData.drop(columns=["employment_type2"])

encodeData["job_title2"] = labelEn.fit_transform(encodeData.job_title)
encodeData.drop(columns=["job_title2"])

encodeData["salary_currency2"] = labelEn.fit_transform(encodeData.salary_currency)
encodeData.drop(columns=["salary_currency2"])

encodeData["employee_residence2"] = labelEn.fit_transform(encodeData.employee_residence)
encodeData.drop(columns=["employee_residence"])

encodeData["company_location2"] = labelEn.fit_transform(encodeData.company_location)
encodeData.drop(columns=["company_location2"])

encodeData["company_size2"] = labelEn.fit_transform(encodeData.company_size)
encodeData.drop(columns=["company_size2"])
encodeData
```

```python
corrDf = encodeData.corr()
stdRatio = 0.3
corrDf
```

```python
feature = list(corrDf.loc[(abs(corrDf.employee_residence2)>stdRatio)&(corrDf.employee_residence2!=1)].index)
feature
```

```python
label = ["employee_residence"]
```

```python
featuredData = encodeData.loc[:,feature]
featuredData
```

```python
minMaxData = featuredData.copy()
```

```python
minMaxData.salary_currency2 = minmax_scale(minMaxData.salary_currency2)
minMaxData.salary_in_usd = minmax_scale(minMaxData.salary_in_usd)
minMaxData.company_location2 = minmax_scale(minMaxData.company_location2)
minMaxData
```

```python
labelData = baseData.loc[:,label]
labelData
```

```python
print(featuredData.shape)
print(labelData.shape)
```

```python
trainDataFeat,\
testDataFeat,\
trainDataLabel,\
testDataLabel = train_test_split(minMaxData, labelData, test_size=0.3, random_state=10)
```

```python
testIndex = testDataLabel.index
```

```python
print(trainDataFeat.shape)
print(trainDataLabel.shape)
print(testDataFeat.shape)
print(testDataLabel.shape)
```

```python
model_mrthod = svm.SVC(random_state=10)
```

```python
model = model_mrthod.fit(X = trainDataFeat,
                         y = trainDataLabel)
```

```python
predict = model.predict(X = testDataFeat)
```

```python
predictData = pd.DataFrame(predict, columns=["predict"])
predictData
```

```python
testDataLabel.reset_index(drop=True, inplace=True)
```

```python
resultData = pd.concat([testDataLabel, predictData], axis=1)
resultData
```

```python
testDataFeat.reset_index(drop=True, inplace=True)
```

```python
finalData = pd.concat([testDataFeat,resultData], axis=1)
finalData
```

```python
finalData.salary_currency2 = featuredData.salary_currency2
finalData.salary_in_usd = featuredData.salary_in_usd
finalData.company_location2 = featuredData.company_location2
finalData
```

```python
accuracy_score(resultData.employee_residence, resultData.predict)
```

```python
print(classification_report(resultData.employee_residence, resultData.predict))
```
